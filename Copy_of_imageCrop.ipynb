```python
import cv2
import matplotlib.pyplot as plt
import easyocr
import torch
import numpy 
import pytesseract
```

```python
def crop_main_image(image):  
  main_image = Image.open(image)

  cropped_main_container = main_image.crop((55,88,1491,1650))

  return cropped_main_container
```

```python
def vertical_crop(image):
  cropped_main_container = crop_main_image(image)
  image_col = []
  left = 0
  right = 375
  
  for i in range(3):
    cropped = cropped_main_container.crop((left, 0, right, 1500))
    left = left + 375
    right = right + 375
    image_col.append(cropped)

  return image_col
```

```python
def horizontal_crop(image):
  image_row = []
  top = 0
  bottom = 149
  for i in range(10):
    cropped = image.crop((0,top,375,bottom))
    top = top + 149
    bottom = bottom + 149
    image_row.append(cropped)

  return image_row
```

```python
def split_image(image):
  image_col = vertical_crop(image)
  split_image_list = []

  for image in image_col:
    columns = horizontal_crop(image=image)

    for column in columns :
      split_image_list.append(column)

  return split_image_list 
```

```python
def ocr(image):
  block_images = split_image(image)
  # reader = easyocr.Reader(['en', 'en'], gpu=False)

  for image in block_images:
      # img = Image.open(io.BytesIO(image))
      gray = cv2.cvtColor(numpy.array(image), cv2.COLOR_BGR2GRAY)
      dilated_img = cv2.dilate(gray, numpy.ones((7, 7), numpy.uint8))
      bg_img = cv2.medianBlur(dilated_img, 21)
      diff_img = 255 - cv2.absdiff(gray, bg_img)
      norm_img = cv2.normalize(diff_img, None, alpha=0, beta=255,norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8UC1)

      work_img = cv2.threshold(norm_img, 0, 255, cv2.THRESH_OTSU)[1]
      # re_shape = cv2.resize(work_img, (1000,500), interpolation = cv2.INTER_AREA)
      # result = reader.readtext(work_img)
      # cv2.imwrite("asda.jpg",re_shape)

      custom_config = r'--oem 3 --psm 6'
      text = pytesseract.image_to_string(work_img, lang='eng', config=custom_config)
     
      print(text)
  # print(image)

```
